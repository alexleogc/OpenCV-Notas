```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### <center> Carregando imagem</center>

É importante notar que o esquema de cores padrão do OpenCV não é o RGB e sim o img_bgr

```python
dir_files = "/home/alexsandrro/Git-projects/OpenCV - Notas/Dados/Imgs-teste/"
```

```python
im_dog_bgr = cv2.imread(dir_files+"pinscher01.jpeg")
plt.imshow(im_dog_bgr)
```

```python
im_dog = cv2.imread(dir_files+"pinscher01.jpeg")
im_dog = cv2.cvtColor(im_dog, cv2.COLOR_BGR2RGB) # Aplicando transformação do espaço de cores BGR para RGB
plt.imshow(im_dog)
```

```python
im_dog_gray = cv2.cvtColor(im_dog, cv2.COLOR_RGB2GRAY) # Convertendo para a escala de cinza
plt.imshow(im_dog_gray,cmap='gray')
```

##### Redimensionando o tamanho das imagens

```python
im_dog.shape # Consultando as dimensões da imagem: 224 pixels em x ; 225 pixels em y ; 3 canais de cores (RGB)
```

```python
im_dog_res = cv2.resize(im_dog,(170,125)) #Redefinindo o número de pixels
plt.imshow(im_dog_res)
```

```python
im_dog_180 = cv2.flip(im_dog,0) #Rotacionar 180º
plt.imshow(im_dog_180)
```

```python
im_dog_esp = cv2.flip(im_dog,1) #Espelhar
plt.imshow(im_dog_esp)
```

```python
im_dog_180_esp = cv2.flip(im_dog,-1) #Rotacionar e espelhar
plt.imshow(im_dog_180_esp)
```
