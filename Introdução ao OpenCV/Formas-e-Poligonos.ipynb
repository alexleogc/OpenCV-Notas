```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_black = np.zeros(shape=(1024,1024,3),dtype=np.int32) #Criando uma imagem de 1024x1024 pixels e 3 canais de cores
img_white = np.ones(shape=(1024,1024,3),dtype=np.int32)*255 #Criando uma imagem de 1024x1024 pixels e 3 canais de cores

#Criando uma imagem 100% vermelha
img_red = np.zeros(shape=(1024,1024,3),dtype=np.int32)*255 #Criando uma imagem de 1024x1024 pixels e 3 canais de cores
img_red[:,:,0] = img_red[:,:,0] + np.asarray(255,dtype=np.int32)

#Criando uma imagem 100% verde
img_green = np.zeros(shape=(1024,1024,3),dtype=np.int32)*255 #Criando uma imagem de 1024x1024 pixels e 3 canais de cores
img_green[:,:,1] = img_green[:,:,1] + np.asarray(255,dtype=np.int32)

#Criando uma imagem 100% azul
img_blue = np.zeros(shape=(1024,1024,3),dtype=np.int32)*255 #Criando uma imagem de 1024x1024 pixels e 3 canais de cores
img_blue[:,:,2] = img_blue[:,:,2] + np.asarray(255,dtype=np.int32)
```

```python
#Plotando as figuras
canvas = ['black','white','red','green','blue']
fig,axes = plt.subplots(ncols=5,figsize=(18,4))
aux=0
for i in canvas:
    axes[aux].set_title(i)
    axes[aux].imshow(eval('img_'+str(i)))
    aux+=1
fig.tight_layout()
```

### Serão explorados os seguintes entes geométricos: Linha, retângulos, círculos, elipses e poligonos

Tutorial utilizando pontos e escalares: https://docs.opencv.org/2.4/doc/tutorials/core/basic_geometric_drawing/basic_geometric_drawing.html

```python
# Criando uma reta
cv2.line(img_white,pt1=(200,100),pt2=(800,800),color=(0,0,0),thickness=20)
plt.imshow(img_white)
```

```python
# Criando uma reta
cv2.line(img_white,pt1=(800,100),pt2=(200,800),color=(255,0,0),thickness=20,lineType=8)
plt.imshow(img_white)
```

```python
# pt1 = Vertice do topo esquerdo ; pt2 = Vertice da base direita
cv2.rectangle(img_white,pt1=(20,200),pt2=(300,600),color=(0,255,255),thickness=10)
plt.imshow(img_white)
```

```python
cv2.rectangle(img_white,pt1=(700,200),pt2=(950,600),color=(0,255,255),thickness=-1) #criando um retângulo preenchido
plt.imshow(img_white)
```

```python
#desenhando um circulo de raio=120 centrado na posição x=500 e y = 800
cv2.circle(img_white,center=(500,800),radius=120,color=(255,0,0),thickness=-1)
plt.imshow(img_white)
```

```python
#desenhando uma elipse de eixo a=200 e eixo b =350 centrado na posição x=500 e y = 150
cv2.ellipse(img_white, center=(500,200), axes=(100,170), angle=0, startAngle= 0, endAngle=360, color=(255,150,50),thickness=-1)
plt.imshow(img_white)
```

#### Polígonos

```python
### Utilizarei a imagem de tela preta

vertices = np.asarray([[100,200],[800,110],[600,850],[400,620]],dtype=np.int32)
vertices
```

```python
vertices = vertices.reshape((-1,1,2))
vertices
```

```python
vertices.shape
```

```python
#desenhando o poligono. O argumento isClosed garante que o poligono será fechado
cv2.polylines(img_black,[vertices],isClosed=True,color=(125,125,0),thickness=20)
plt.imshow(img_black)
```

```python
#desenhando o poligono preenchido.
cv2.fillPoly(img_black,[vertices],color=(125,125,125))
plt.imshow(img_black)
```
